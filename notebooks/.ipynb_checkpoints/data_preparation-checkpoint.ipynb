#### Clean Libraries

```python
# Create function to deal properly capitalize words in text columns and to remove leading and trailing spaces
import re

def clean_text_column(df, column_name):
    """
    Cleans a text column in a DataFrame by:
    - Stripping leading and trailing spaces
    - Converting text to lowercase and then title case
    - Removing double spaces
    
    Parameters:
    - df: pandas DataFrame
    - column_name: str, the name of the column to clean
    
    Returns:
    - df with cleaned text column
    """

    # Handle NaN values by converting them to empty strings
    df[column_name] = df[column_name].fillna('').str.strip().str.lower().str.title()
    
    # Remove double (or more) spaces
    df[column_name] = df[column_name].apply(lambda x: re.sub(r'\s+', ' ', x))
    
    return df
```

```python
# Let's test it
clean_text_column(libraries, 'name')
```

```python
# Apply function to relevant columns in libraries dataframe
libraries = clean_text_column(libraries, 'name')
libraries = clean_text_column(libraries, 'city')
libraries = clean_text_column(libraries, 'region')

libraries.head()
```

```python
# Next, let's create function that cleans postal code
def clean_postal_code(df, column_name):

    # Remove non-digit characters
    df[column_name] = df[column_name].astype(str).str.replace(r'\D', '', regex=True)
    # Convert to integer, handling NaNs
    df[column_name] = pd.to_numeric(df[column_name], errors='coerce').astype('Int64')
    
    return df

libraries = clean_postal_code(libraries, 'postal_code')
libraries.head()
```

For some rows, we have postal code but city or region (or both) are emtpy. To remedy this, we can download free database of zipcodes and insert missing data by using what we have.

Dataset we can use for this purpose can be found on this link https://github.com/scpike/us-state-county-zip/blob/master/geo-data.csv

```python
# Download and store geo-data.csv
import requests

# URL of the CSV file
url = "https://raw.githubusercontent.com/scpike/us-state-county-zip/master/geo-data.csv"
save_path = path+'geo_data.csv'

# Download and save the file
response = requests.get(url)
with open(save_path, "wb") as file:
    file.write(response.content)

# Store dataset for potential future use
geo_data = pd.read_csv(path+'geo_data.csv')

# Rename column names in geo_data dataframe to avoid conflicts
geo_data = geo_data[['state', 'zipcode', 'city']]
geo_data.columns = ['state_geo', 'zipcode', 'city_geo']
```

```python
# Rename Or with Oregon
libraries.loc[libraries['region']=='Or', 'region'] = 'Oregon'

# Convert postal_code to string so we can join it with geo data
libraries['postal_code'] = libraries['postal_code'].apply(lambda x: str(x).split('.')[0])
# Add geo data
libraries = pd.merge(libraries, geo_data, left_on='postal_code', right_on='zipcode', how='left')

# Input missing cities and states for libraries where we have zipcodes
libraries.loc[(libraries['city'].isna() | libraries['city']=='') & ~libraries['city_geo'].isna(), 'city'] = libraries['city_geo']
libraries.loc[(libraries['region'].isna() | libraries['region']=='') & ~libraries['state_geo'].isna(), 'region'] = libraries['state_geo']

# Let's also convert street addreses to have the same format
libraries["street_address"] = libraries["street_address"].astype(str).apply(lambda x: ' '.join(x.split()).title())
```

#### Clean Checkouts

In this dataset, some of the entries do not have id and a lot of them have wrong date format or obviously wrong entry (i.e. date_checkout = 1819-04-05)

By looking at this dataset, we can observe that, by removing erroneous dates, we would remove 340+ rows out of 2000, which is roughly 17% of the table. That is unnacceptable so we should find some way to clean date columns as much as we can.

Some of the examples of different entries (formats) are following:
- 14-09-20
- 20180329%
- 2018 07 06
- 2018|05|13
- 1800-12-07

First format is prevalent but in order to include dates like the last example, we should try to convert all other formats to have the format as the last one.

Additionaly, we should figure out what to do with rows where no dates are present. Number of rows which have missing dates number about 6% of the dataset. We might decide to not include these entries in our analysis or we might assume that these checkout are late as well. We should analyze data with both assumptions and verify if those entries change conclusion.

Simple solution would be to imput missing date_returned with current data but we might want to omit rows where there are no date_checkout entries.

```python
# Let's create function that cleans dates as we will need to apply simmilar logic several times
def clean_dates(df, column_name):
    
    # Remove non-numeric characters (like '%', '|', etc.) except valid date delimiters
    df[column_name] = df[column_name].str.replace(r'[^0-9\- ]', '', regex=True)
    
    # Convert to datetime format with multiple possible formats
    df[column_name] = pd.to_datetime(df[column_name], errors='coerce', dayfirst=False)
    
    # Format the output as 'YYYY-MM-DD'
    df[column_name] = df[column_name].dt.strftime('%Y-%m-%d')

    # Convert column type to date
    df[column_name] = pd.to_datetime(df[column_name], errors='coerce')

    return df

# Clean two dates columns in checkouts dataframe
checkouts = clean_dates(checkouts, 'date_checkout')
checkouts = clean_dates(checkouts, 'date_returned')

# Add todays date to missing entries in date returned
from datetime import datetime

# Get today's date in the desired format
today = datetime.today().strftime('%Y-%m-%d')

# Replace NaN entries with today's date
checkouts['date_returned'] = checkouts['date_returned'].fillna(today)

# And while we are at this dataframe, let's calculate difference between date_returned and date_checkout in number of days
# Calculate the difference in days
checkouts['diff_days'] = (checkouts['date_returned'] - checkouts['date_checkout']).dt.days

checkouts.head(10)
```

```python
# Function to correct unrealistic future dates
def correct_future_dates(date):
    if pd.isnull(date):
        return None
    if date.year >= 2100:
        return date.replace(year=date.year - 100)
    return date

checkouts["date_checkout"] = checkouts["date_checkout"].apply(correct_future_dates)
checkouts["date_returned"] = checkouts["date_returned"].apply(correct_future_dates)
```

#### Checkouts Observations
Numerous rows have negative date difference between checkout and return. This is clearly wrong. 

The problem is, these entries consist more than 15% of dataset. One assumption is that the dates are reversed.

Some of the entries for date_checkout are from 1800s. One of the ideas is to join this dataset with books dataset and then set date_checkout dates so that they are no lower than publish date of rented books.

In any case, later on, additional consideration is necessary when analyzing these dates.

Some of the rows in city and state columns are empty. Luckily, we have all zipcodes which we can use to get proper city and state names.

Before we insert missing entries, we can notice that zipcodes in the customers dataset have 6 digits while zipcodes should have only 5. If we remove trailing zero, we will get proper zipcode which we can use to join customers dataset with geo_data dataset.

```python
# Remove trailing zero from zipcodes
# Function to remove one trailing zero
def remove_trailing_zero(zipcode):
    
    if isinstance(zipcode, str) and zipcode.endswith("0"):
        # Remove the last character
        return zipcode[:-1]
        
    return zipcode

# Apply the function to the 'zipcode' column
customers["zipcode"] = customers["zipcode"].astype(str).apply(remove_trailing_zero)

# Join customers with geo data
customers = pd.merge(customers, geo_data, on='zipcode', how='left')
```

```python
# Replace nan's in customer city and state with geo data based on zipcode
customers.loc[customers['city'].str.len()==0 & ~customers['city_geo'].isna(), 'city'] = customers['city_geo']
customers.loc[customers['state'].str.len()==0 & ~customers['state_geo'].isna(), 'state'] = customers['state_geo']

# Now, we can remove columns we don't need anymore
customers.drop(columns=['state_geo', 'city_geo'], inplace=True)
```

```python
# Let's calculate age based on birth date
# Convert birth_date to datetime format
customers['birth_date'] = pd.to_datetime(customers['birth_date'], errors='coerce')
# Calculate age
customers['age'] = customers['birth_date'].apply(lambda x: datetime.now().year - x.year if pd.notna(x) else None)

customers.head()
```

#### Clean Books

We can notice that price and pages columns contain erroneous entries. We need to clean those columns and to convert them to proper type.

Also, categories column contains entries in list format. For easier analysis, we will remove uneccessary characters. Same thing with authors column.

Regarding column publisher, more than 60% of rows are empty so it's not realy usable.

Finally, published_date contains various date types so we need to format all rows to be of the same format.


```python
# Function to clean and convert prices to float
def clean_price(value):

    # Ensure it's a string
    if isinstance(value, str):  
        # Remove non-numeric characters except "."
        value = re.sub(r"[^\d.]", "", value)
    try:
        # Convert to float
        return float(value)
        
    except ValueError:
        # Return None if conversion fails
        return None

# We can use same function on both price and pages columns
books['price'] = books['price'].apply(clean_price)
books['pages'] = books['pages'].apply(clean_price).astype('int')
```

```python
# Now, let's clean special characters from text columns
# Function to clean special characters
def clean_text(value):

    # Ensure it's a string
    if isinstance(value, str):
        
        # Keep only letters, numbers, and spaces
        return re.sub(r"[^a-zA-Z0-9\s]", "", value)
        
    return value

# Let's remove special characters
books['authors'] = books['authors'].apply(clean_text)
books['categories'] = books['categories'].apply(clean_text)

# Now apply function for cleaning text we already have
books = clean_text_column(books, 'authors')
books = clean_text_column(books, 'categories')
```

```python
# Function to convert different date formats to YYYY-MM
def convert_to_yyyy_mm(date_str):
    # Ensure it's a string
    if isinstance(date_str, str):
        
        # Case 1: YYYY (only year)
        if re.fullmatch(r"\d{4}", date_str):
            # Default to January
            return f"{date_str}-01"

        # Case 2: YYYY-MM (year and month)
        elif re.fullmatch(r"\d{4}-\d{2}", date_str):
            # Reformat to YYYY-MM
            return f"{date_str[:4]}-{date_str[-2:]}"

        # Case 3: YYYY-MM-DD (ignore day, assume YY is YYYY)
        elif re.fullmatch(r"\d{4}-\d{2}-\d{2}", date_str):
            try:
                # Parse as datetime and extract MM-YYYY
                parsed_date = datetime.strptime(date_str, "%Y-%m-%d")
                return parsed_date.strftime("%Y-%m")
            except ValueError:
                # Handle parsing errors
                return None  

    # Return None if format is unrecognized
    return None

# Rename column to be in accordance with naming convention
books.rename(columns={"publishedDate": "published_date"}, inplace=True)

# Apply the function to the date column
books["published_date"] = books["published_date"].apply(convert_to_yyyy_mm)
```

### Join Data

In this step we will need to figure out how to join datasets and what kind of relations they have (many to many, one to many, etc.).

Ideally, we should aim to create one dataset with all data in it so it's easier to analyze it. This isn't always good idea but in this case I believe it is feasible.

```python
# It turns out that we can join all data in one dataframe
# Add prefixes to dataframe column names to keep track of sources of data
books.columns = ['bk_'+col for col in books.columns]
checkouts.columns = ['chk_'+col for col in checkouts.columns]
customers.columns = ['cust_'+col for col in customers.columns]
libraries.columns = ['lib_'+col for col in libraries.columns]

# Join checkouts and libraries
df = pd.merge(checkouts, libraries, left_on='chk_library_id', right_on='lib_id', how='left')

# Add customer info
df = pd.merge(df, customers, left_on='chk_patron_id', right_on='cust_id', how='left')

# Add books info
df = pd.merge(df, books, left_on='chk_id', right_on='bk_id', how='left')
```

```python
# Now that we have joined all datasets, we can remove obsolete columns
obs_cols = ['chk_id', 'chk_patron_id', 'chk_library_id', 'lib_id', 'lib_postal_code',
            'cust_id', 'cust_zipcode', 'bk_id']
df = df.drop(columns=obs_cols)

# We can save this prepared dataset
df.to_csv(path+'df.csv', index=False)
```
